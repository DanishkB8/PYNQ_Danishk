# Image Resizing using PIL

This notebook demonstrates how to resize images using the Python Imaging Library (PIL). We will read an image, display its original size, resize it by a given factor, and display the resized image.

## Step 1: Import Required Libraries


```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Step 2: Load and Display the Original Image
Load the image to be resized and display its dimensions and content.

```python
image_path = "pynq.jpeg"
original_image = Image.open(image_path)
```

```python
canvas = plt.gcf()
size = canvas.get_size_inches()
canvas.set_size_inches(size*2)

old_width, old_height = original_image.size
print("Image size: {}x{} pixels.".format(old_width, old_height))
_ = plt.imshow(original_image)
```

## Step 3: Set Resize Factor and Compute New Dimensions
Define the resize factor and calculate the dimensions of the resized image.

```python
resize_factor = 4
new_width = int(old_width/resize_factor)
new_height = int(old_height/resize_factor)
```

## Step 4: Resize the Image
Resize the image using the specified resize factor and display the resized image.

```python
resized_image = original_image.resize((new_width, new_height), Image.BILINEAR)
```

```python
print("Image size: {}x{} pixels.".format(new_width, new_height))
_ = plt.imshow(resized_image)
```

 ## Step 5: Measure Resize Execution Time
Use the timeit magic command to measure the time taken to resize the image.

```python
%%timeit
resized_image = original_image.resize((new_width, new_height), Image.BILINEAR)
```
